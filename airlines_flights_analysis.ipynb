```python
#Q.2. Show Bar Graphs representing the Departure Time & Arrival Time.


plt.figure(figsize = (16,4))

plt.subplot(1,2,1)

plt.bar(df['departure_time'].value_counts().index, df['departure_time'].value_counts().values)

plt.title("Departure Time")

plt.xlabel("Day Time")

plt.ylabel("Day Frequency")

plt.subplot(1,2,2)

plt.bar(df['arrival_time'].value_counts().index, df['arrival_time'].value_counts().values, color = ["r"])

plt.title("Arrival Time")

plt.xlabel("Day Time")

plt.ylabel("Day Frequency")

plt.show()
```

```python
#Q.3. Show Bar Graphs representing the Source City & Destination City.

plt.figure(figsize = (16,4))

plt.subplot(1,2,1)

plt.barh(df['source_city'].value_counts().index, df['source_city'].value_counts().values)

plt.title("Source City Number of flights")

plt.ylabel("City")

plt.xlabel("Number of flights")

plt.subplot(1,2,2)

plt.barh(df['destination_city'].value_counts().index, df['destination_city'].value_counts().values, color = ['r'])

plt.title("Destination City Number of flights")

plt.ylabel("City")

plt.xlabel("Number of flights")

plt.show()
```
